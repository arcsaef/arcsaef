```python
# on hold rojects are not included
for prsn in ppl_saef:
    p = ppl_saef.get(prsn)
    # Active members, excl. Governance Committee memebers
    if p['State'] == 'Active' and p['Position'] != 'Advisory':
        if len(p['Projects']) > 0:
            projects = f"{p['Projects']}"
            pcl = re.sub(r" \(\d+\)", "", projects)
            pcl = re.sub(r" \(\)", "", pcl)
            pcl = re.sub(r"\[", "", pcl)
            pcl = re.sub(r"\]", "", pcl)
            pcl = re.sub(r"\'", "", pcl)
            pcl = re.sub(r" ", "", pcl)
            pcl = re.sub(r",", ", ", pcl)
            cl = f"{p['Title']}; {p['FirstName']}; {str.strip(p['LastName'] + ' ' + p['Postnominal'])}; {p['Email']}; {p['Organisation']}; {p['Position']}; {pcl}"
            print(cl)         
        else:
            print(f"{p['Title']}; {p['FirstName']}; {str.strip(p['LastName'] + ' ' + p['Postnominal'])}; {p['Email']}; {p['Organisation']}")
```
